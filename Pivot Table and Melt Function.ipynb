# Melt()

```python
import pandas as pd
va = pd.DataFrame({"days":[1,2,3,4,5,6],"eng" : [10,12, 14, 15, 16, 12], "maths" : [17,18,19, 13, 14,16]})
va
```

```python
# pd.melt(var)
# pd.melt(va,id_vars=["days"],var_name="python")
pd.melt(va,id_vars=["days"],var_name="python",value_name="wscube")
```

# Pivot()

```python
import pandas as pd
# var = pd.DataFrame({"days":[1,2,3,4,5,6],
#                     "st_name":["a","b","c","a","b","c"],"eng" : [10,12, 14, 15, 16, 12], "maths" : [17,18,19, 13, 14,16]})
var = pd.DataFrame({"days":[1,1,1,1,2,2],
                    "st_name":["a","b","a","a","b","b"],"eng" : [10,12, 14, 15, 16, 12], "maths" : [17,18,19, 13, 14,16]})
var 
```

```python
# var.pivot(index="days",columns="st_name")
var.pivot(index="days",columns="st_name",values="eng")
```

```python
# var.pivot_table(index="st_name",columns="days",
#                 aggfunc="sum")
var.pivot_table(index="st_name",columns="days",
                aggfunc="mean",margins="True")
```
